```python
import pandas as pd, numpy as np
from scipy import sparse
from sklearn.metrics.pairwise import pairwise_distances, cosine_similarity
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df = pd.read_csv('CSV\'s/full_whiskey_list.csv')
```

```python
df.head()
```

```python
df['rating'].value_counts()
```

```python
mms = MinMaxScaler()
df['rating'] = mms.fit_transform(df['rating'].values.reshape(-1,1))
```

```python
pivot_df = df.pivot_table(index = 'name', columns = 'type', values='rating')
```

```python
pivot_df
```

```python
pivot_df_sparse = sparse.csr_matrix(pivot_df.fillna(0))
```

```python
print(pivot_df_sparse)
```

```python
# pairwise_recommender = pairwise_distances(pivot_df_sparse, metric='cosine')
```

```python
cs_recommender = cosine_similarity(pivot_df_sparse)
```

```python
# pairwise_whiskey_recommender = pd.DataFrame(pairwise_recommender,
#                                   index=pivot_df.index,
#                                   columns=pivot_df.index)
```

```python
cs_whiskey_recommender = pd.DataFrame(cs_recommender,
                                     index=pivot_df.index,
                                     columns=pivot_df.index)
```
