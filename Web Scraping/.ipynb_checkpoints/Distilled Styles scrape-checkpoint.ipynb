```python
import pandas as pd, numpy as np, requests
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
url = 'https://distiller.com/search'
driver = webdriver.Chrome(executable_path="../chromedriver/chromedriver")
```

```python
res = requests.get(url)
driver.get(url)
```

```python
res.status_code
```

```python
soup = BeautifulSoup(driver.page_source)
```

```python
styles = pd.DataFrame(columns=['style/type'])
```

```python
whiskey_urls = []
page_num = np.array(range(1,345))
for i in page_num:
    whiskey_url = url + '?page=' + i.astype(str)
    whiskey_urls.append(whiskey_url)
whiskey_urls
```

```python
for url in whiskey_urls:
    driver.get(url)
    soup = BeautifulSoup(driver.page_source)
    table = soup.find('ol', attrs={'class':'spirits'})
    for row in table.find_all('li'):
        for a in row.find_all('a'):
            try:
                style = a.find('p', {'class': 'accented-text origin'}).text
                styles.loc[len(styles)] = [style]
            except:
                pass            
```

```python
styles.to_csv('../CSV\'s/distilled_styles.csv', index = False)
```
