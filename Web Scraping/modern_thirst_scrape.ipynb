```python
import pandas as pd, requests, time
from bs4 import BeautifulSoup
from selenium import webdriver
from random import randrange
```

```python
url = 'https://modernthirst.com/home/bourbon-and-whiskey-reviews/'
driver = webdriver.Chrome(executable_path="chromedriver/chromedriver")
```

```python
res = requests.get(url)
driver.get(url)
```

```python
soup = BeautifulSoup(driver.page_source)
```

```python
res.status_code
```

```python
whiskey2 = pd.DataFrame(columns = ['name', 'rating'])
```

```python
tbody = soup.find('tbody', attrs={'class':'row-hover'})
whiskey2 = pd.DataFrame(columns = ['name', 'rating'])
for row in tbody.find_all('tr', attrs = {'role':'row'}):
    time.sleep(randrange(4,10))
    name = row.find('td', {'class':'column-1'}).text
    rating = row.find('td', {'class':'column-5'}).text
    whiskey2.loc[len(whiskey2)] = [name, rating]

whiskey2
```
